# Projeto de Regressão Múltilpla usando Redes Multilayer Perceptron
- O projeto abaixo se baseia na análise estatística e então modelagem da arquitetura de uma rede neural com 3 layers.

```python
import pandas as pd
from datetime import datetime
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 18, 8
import numpy as np
```

### Dados
- Os dados aqui utilizados são medições do nível de água da barragem do município Sobradinho-BA. As medições compreendem de 1931 a 2015, dentre os meses de Janeiro a Dezembro.

```python
data = pd.read_csv('SOBRADINHO_w_dates.csv', sep=';')
data[:5]
```

### Transformação para Série Temporal
- Abaixo seguem transformações para que os dados fiquem no formato de uma série temporal (ST), a fim de facilitar o posterior tratamento dos mesmos. 
- Um conjunto de dados é dito ser uma série quando os dados seguem uma ordem. Aqui essa ordem  é cronológica.

```python
df = {'DATE':[],'MEASURE':[]}
columns = data.keys()
for row in range(len(data)):
    for i,column in enumerate(columns):
        if i==0:
            pass
        else:
            df['DATE'].append(datetime(data[columns[0]][row],i,1))
            df['MEASURE'].append(data[column][row])
df = pd.DataFrame(data=df)
df[:5]
```

### Estacionaridade
- Uma ST é dita estacionária quando suas propriedades estatísticas, como a média e a variância, permanecem constantes ao longo do tempo. Verificar sua estacionaridade significa observar o comportamento específico ao longo do tempo. No gráfico a seguir, os dados foram traçados e analisados visualmente. Verificou-se uma certa tendência ao longo do tempo juntamente com algumas variações sazonais.

```python
df.set_index('DATE').plot()
plt.show()
```

- Ao plotar o desvio padrão, fica claro que o mesmo apesar de variar muito ao longo do tempo, mantêm uma certa estabilidade. Ou seja, percebemos aqui que a série é estacionária. Afim de certirficar-nos disso, iremos realizar o teste de estacionaridade.


```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.DataFrame.rolling(timeseries,window=12,center=False).mean()
    rolstd = pd.DataFrame.rolling(timeseries,window=12,center=False).std()
    #Plot rolling statistics:
    plt.figure(figsize=(18,8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries['MEASURE'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
test_stationarity(df[:504].set_index('DATE'))
```

- Primeiramente, foi feita uma transformação logarítmica dos dados. O objetivo aqui era penalizar valores altos.
- Em seguida, sentiu-se a necessidade de identificar a presença de ruído afim de removê-lo. Para isso, foi utilizado a técnica das médias.

```python
df_log = pd.DataFrame({'DATE':df['DATE'],'MEASURE':np.log(df['MEASURE'])}).set_index('DATE')
plt.figure(figsize=(18,8))
plt.plot(df_log)
```

- Aqui foram projetadas as médias dos anos, ou seja, médias do 12 meses de cada ano. Abaixo a linha vermelha mostra esse comportamento.

```python
moving_avg = pd.DataFrame.rolling(df_log,window=12,center=False).mean()
plt.figure(figsize=(18,8))
plt.plot(df_log)
plt.plot(moving_avg, color='red')
```

- O valores encontrados foram então extraídos da série original. Observa-se que, uma vez que foi tomado a média dos últimos 12 valores, a média não está definida para os 11 primeiros valores.
- O valor crítico encontrado foi de  5%, ou seja, é possível dizer com 95% de confiança que esta é uma série estacionária. Foi utilizado então uma "média móvel ponderada", onde os valores mais recentes recebem um peso maior. A escolhida foi a média móvel ponderada exponencialmente, onde os pesos são atribuídos a todos os valores anteriores com um fator de decaimento.


```python
df_log_moving_avg_diff = df_log - moving_avg
df_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(df_log_moving_avg_diff)
```

- Aqui foi definido a quantidade de decaimento exponencial, mas apenas como suposição.Foi então removido isso da série para conferir a estacionariedade.

```python
expwighted_avg = pd.DataFrame.ewm(df_log, halflife=12, ignore_na=False,min_periods=0,adjust=True).mean()
plt.figure(figsize=(18,8))
plt.plot(df_log)
plt.plot(expwighted_avg, color='red')
```

- A ST tem variações ainda menores em média e desvio padrão em magnitude. Além disso, a estatística de teste é menor do que o valor crítico de 1%, o que é melhor do que o caso anterior. 

```python
df_log_ewma_diff = df_log - expwighted_avg
test_stationarity(df_log_ewma_diff)
```

- O próximo passo foi remover a tendência e a sazonalidade. Um dos métodos mais comuns para lidar com a tendência e a sazonalidade é a diferenciação. Nesta técnica, é tomado a diferença da observação em um instante particular com a do instante anterior. Isso funciona principalmente bem na melhoria da estacionança. A diferenciação de primeira ordem então foi aplicada.

```python
df_log_diff = df_log - df_log.shift()
plt.figure(figsize=(18,8))
plt.plot(df_log_diff)
```

- A técnica aparenta ter reduzido consideravelmente a tendência, como é mostrado a seguir. Observa-se que a estatística de teste é inferior ao valor crítico de 10%, portanto, o ST está parado com 90% de confiança.

```python
df_log_diff.dropna(inplace=True)
test_stationarity(df_log_diff)
```

- Aqui é possível ver que a tendência e a sazonalidade foram separadas dos dados tornando possível a modelagem dos resíduos.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_log)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(18,8))
plt.subplot(411)
plt.plot(df_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

- Abaixo a estacionaridade dos resíduos é verificada.A estatística de teste Dickey-Fuller é significativamente menor do que o valor crítico de 1%. Então, esta ST está muito perto de estacionário. 

```python
df_log_decompose = residual
df_log_decompose.dropna(inplace=True)
test_stationarity(df_log_decompose)
```

- Os gráficos da Função de autocorrelação e da Função de Autocorrelação Parcial após a diferenciação são mostrados abaixo. As duas linhas pontilhadas em ambos os lados de 0 são os intervalos de confiança. Estes podem ser usados para determinar os valores de atraso em que os gráficos cruzam o intervalo de confiança superior pela primeira vez. Em ambos os gráficos esse valor está perto de 2.5. 


```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(df_log_diff, nlags=20)
lag_pacf = pacf(df_log_diff, nlags=20, method='ols')
#Plot ACF: 
plt.figure(figsize=(18,8))
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

# Matriz de Correlação
- Em mãos dos dados, iremos agora construir a matiz de correlação, onde poderemos observar uma relação entre os dados, para assim podermos definir a quantidade de meses que poderiam ser usados na predição.

```python
import keras 
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import Sequential 
from keras.layers import Dense, Dropout 
from keras.optimizers import SGD, RMSprop, Adam
from keras.layers.advanced_activations import PReLU, LeakyReLU
from keras.callbacks import CSVLogger, EarlyStopping, LearningRateScheduler
from math import exp
# Loading data
data = pd.read_csv('SOBRADINHO.csv')
correl = []
for y in range(0,len(data)):
    for x in data.keys():
        correl.append(data[x][y])
df = {'month_T-10':[],'month_T-9':[],'month_T-8':[],'month_T-7':[],
      'month_T-6':[],'month_T-5':[],'month_T-4':[],'month_T-3':[],
      'month_T-2':[],'month_T-1':[],'month_T-0':[],'month_T+1':[]}
while len(correl) > 12:
    df['month_T-10'].append(correl[0])
    df['month_T-9'].append(correl[1])
    df['month_T-8'].append(correl[2])
    df['month_T-7'].append(correl[3])
    df['month_T-6'].append(correl[4])
    df['month_T-5'].append(correl[5])
    df['month_T-4'].append(correl[6])
    df['month_T-3'].append(correl[7])
    df['month_T-2'].append(correl[8])
    df['month_T-1'].append(correl[9])
    df['month_T-0'].append(correl[10])
    df['month_T+1'].append(correl[11])
    del correl[0]
df = pd.DataFrame(df,columns=['month_T-10','month_T-9','month_T-8','month_T-7',
                              'month_T-6','month_T-5','month_T-4','month_T-3',
                              'month_T-2','month_T-1','month_T-0','month_T+1'])
# Correlation Heatmap
plt.figure(figsize=(18,8))
sns.heatmap(df.corr(), annot=True, fmt=".2f")
plt.show()
```

# Modelagem da Rede Neural
- Nessa parte, começaremos a modelagem da rede neural. Afim de otimizar a rede, procuramos técnicas e formas de deixar mais simples todo o sistema. Essa parte também compreende o tratamento dos dados para cada tipo de rede de acordo com seu número de entradas.

### Dos dados
- Antes dos dados serem usados, eles antes tratados de forma que sejam separados e usados de acordo com a nescessidade da rede. 
- Foram divididos pelo método de validação cruzada: Treino (50%), Teste (25%) e Validação (25%).
- Foram normalizados entre -1 e 1, afim de se adaptar melhor às funções de ativação aqui escolhidas (vide próximo tópico)

### Optimizadores
- Afim de construir um modelo forte e robusto, decidimos usar o otimizador Stochastic Gradient Descent
- Com o intuito de evitar overfitting, implementamos o LearningSchedule, que é uma função que monitora a variável escolhida (no nosso caso, validation_loss, cujo objetivo da rede neural é minimiza-lo) e cessa a RN quando a variável validation_loss volta a aumentar.
- Implementamos também uma função que diminui a taxa de aprendizagem exponencialmente ao longo das épocas, assim ajudando a rede a convergir.

### Métricas
- As métricas utilizadas, foram: Loss, Validation Loss, Mean Average Error, Validation Mean Average Error.

```python
#on_epoch_end: logs include acc and loss, and optionally include val_loss (if validation is enabled in fit), 
#              and  val_acc (if validation and accuracy monitoring are enabled).
#on_batch_begin: logs include size, the number of samples in the current batch.
#on_batch_end: logs include loss, and optionally acc (if accuracy monitoring is enabled).
class EPOHistory(keras.callbacks.Callback):
  def on_train_begin(self, logs={}):
    self.epochs = []
    self.mae_epoch = []
    self.losses_epoch = []
    self.val_mae_epoch = []
    self.val_losses_epoch = []
    self.lr = []
  def on_epoch_end(self, epoch, logs={}):
    self.epochs.append(epoch)
    self.mae_epoch.append(logs.get('mean_absolute_error'))
    self.losses_epoch.append(logs.get('loss'))
    self.val_mae_epoch.append(logs.get('val_mean_absolute_error'))
    self.val_losses_epoch.append(logs.get('val_loss')) 
    self.lr.append(exp_decay(len(self.losses_epoch)))

def exp_decay(epoch):
   initial_lrate = 0.1
   k = 0.1
   learning_rate = initial_lrate * exp(-k*epoch)
   return learning_rate
    
def fitModel(monthsSize = 3, forecastSize = 1, activationFunction='tanh', hiddenLayer=13, epochsModel=400):
  raw_data = []
  keys = data.keys()
  for row in xrange(0,len(data)):
      for key in keys:
          raw_data.append(data[key][row])
  data_mean = np.mean(raw_data)
  den = np.max(raw_data) - float(np.min(raw_data))
  for cell in xrange(0,len(raw_data)):
    raw_data[cell] = (raw_data[cell] - data_mean)/(den)
  arraySize = monthsSize+forecastSize
  batched_data = []                                   #array that will receive the data batched with lenght of the arraySize
  while len(raw_data) > arraySize:
      batched_data.append(raw_data[:arraySize])
      del raw_data[0]
  names = ['month-11','month-10','month-9',\
           'month-8','month-7','month-6',\
           'month-5','month-4','month-3',\
           'month-2','month-1','month_0','month+1']
  namesArray = names[-arraySize:]
  dict_data = {el:[] for el in namesArray}
  for row in xrange(0,len(batched_data)):
      for i,name in enumerate(namesArray):
          dict_data[name].append(batched_data[row][i])
  df = pd.DataFrame(dict_data)
  df = df.sample(frac=1).reset_index(drop=True)         #shuffle
  source_data = [[df[x][y] for x in namesArray if x!=namesArray[-1]] for y in xrange(0,len(df))]
  source_labels = [[df[namesArray[-1]][x]] for x in xrange(0,len(df))]
  funfundzwanzig = int(round(len(df)/4.0))
  validation_data = source_data[:funfundzwanzig]
  validation_labels = source_labels[:funfundzwanzig]  
  train_data = source_data[funfundzwanzig:-funfundzwanzig]
  train_labels = source_labels[funfundzwanzig:-funfundzwanzig]
  test_data = source_data[-funfundzwanzig:]
  test_labels = source_labels[-funfundzwanzig:]
  model = Sequential()
  if activationFunction=='LeakyReLU':
    model.add(LeakyReLU(alpha=0.3, input_shape=(monthsSize,)))
  elif activationFunction=='PReLU':
    model.add(PReLU(alpha_initializer='glorot_normal', alpha_regularizer=None, alpha_constraint=None, shared_axes=(0,1),input_shape=(monthsSize,)))
  else:
    model.add(Dense(hiddenLayer, activation=activationFunction, input_shape=(monthsSize,))) 
  model.add(Dense(forecastSize, activation='linear'))
  learning_rate = LearningRateScheduler(exp_decay)
#   decay_rate = learning_rate/epochsModel
  momentum = 0.8
  opt = SGD(lr=0.01, momentum=momentum, decay=0.0, nesterov=False)
#  opt = RMSprop(lr=learning_rate, rho=0.9, epsilon=1e-08, decay=0.0)
#  opt = Adam(lr=learning_rate, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
  model.compile(loss='mean_squared_error', 
                optimizer=opt, 
                metrics=['mae'])
  epochsHistory = EPOHistory()
  earlyStopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=300, verbose=0, mode='auto')
  history = model.fit(train_data, train_labels, 
                      batch_size=3, 
                      epochs=epochsModel, 
                      verbose=0, 
                      validation_data=( validation_data, validation_labels), 
                      callbacks=[epochsHistory, earlyStopping, learning_rate])
  score = model.evaluate( test_data, test_labels, verbose=0)
  print('[INFO] Epochs ---------------------> %s'%(epochsHistory.epochs[-1]))
  print('[INFO] Test loss ------------------> %0.7f'%(score[0]))
  print('[INFO] Test mean squared error ----> %0.6f%%'%(score[1]*100.0))
  return score, model, history, epochsHistory
```

# Arquitetura da rede neural
### Número de entradas
- Através da análise da matriz de correlação, podemos inferir que os meses que possuem uma relação linear com o mês a ser previsto começam a partir do 2º mês antes do previsto, significando então que a entrada da rede poderiam ser até de 2 meses antes. Entretanto, observamos que entre os meses T-2 e T-4 a relação com o mês a ser previsto começa a se tornar inversamente proporcional, para então a partir de T-10 a correlação voltar a ser proporcional. Isso significa que houveram relações não lineares ao longo dos meses, o que faz sentido dado que vimos anteriormente que os dados se comportam sazonalmente. Para a decisão de quais meses entrarão no treino, decidimos que seria: [1, 2, 3, 6, 12].
### Número de neurônios da camada escondida
- Em mãos de várias fontes, a melhor conclusão que poderíamos chegar é a soma das entradas + a saída = número de neurônios na camada escondida, para então testar adicionando de 5 em 5. Como possuímos entradas variáveis de 1 a 12, decidimos testar várias combinações diferentes, como: [5, 12, 16, 24, 36].
### Número de saídas
- Afim de atender os requisitos do projeto, o número de saídas da rede será de apenas 1, significando assim que a rede irá apenas prever o próximo mês em relação aos meses de entrada.
### Funções de ativação
- As funções de ativações aqui utilizadas são as mais famosas e possuem diferentes dinâmicas porém todas pertinentes ao modelo de regressão. Por isso, nos pareceu interessante utilizar as seguintes funções: Parametric Rectified Linear Unit - PReLU, Leaky Rectified Linear Unit - LeakyReLU, Tangente Hiperbólica, Rectified Linear Unit - ReLU.
- Para a função de ativação entre a camada escondida e a camada de saída, foi escolhida a função linear (dado que era o mais recomendado em problemas de regressão com redes neurais).

```python
# Variables
monthsSize = 2
forecastSize = 1

months_Size = [1, 2, 3, 6, 12]
forecast_Size = [1]
activation_Function=['PReLU', 'LeakyReLU','tanh','relu']
hidden_Layer=[5, 12, 16, 24, 36]
epochs_Model=[600]

# combinations = [[months_Size[a],forecast_Size[b],activation_Function[c],hidden_Layer[d],epochs_Model[e]] for a in xrange(0,len(months_Size)) for b in xrange(0,len(forecast_Size)) for c in xrange(0,len(activation_Function)) for d in xrange(0,len(hidden_Layer)) for e in xrange(0,len(epochs_Model))]
combinations = [[months_Size[a],forecast_Size[b],activation_Function[c],hidden_Layer[d],epochs_Model[e]]\
                for a in range(0,len(months_Size))\
                for b in range(0,len(forecast_Size))\
                for c in range(0,len(activation_Function))\
                for d in range(0,len(hidden_Layer))\
                for e in range(0,len(epochs_Model))]
```

# Treino
- Iremos agora, através da técnica de grid search, testar vários parâmetros simultaneamente.

```python
grid_search = []
for i,params in enumerate(combinations):
  grid_search.append(fitModel(*params))
  print "[INFO] Model %s trained of %s, progress:  %0.2f%%\n"%(i+1,len(combinations),(100.0*i/len(combinations)))
```

# Resultados
- Ordenaremos então os resultados de acordo com o menor MAE, afim de selecionarmos os melhores modelos para análise.

```python
grid_list = []
for x in range(0,len(grid_search)):
    grid_list.append(list(grid_search[x]))
for x in range(0,len(grid_list)):
    grid_list[x].append(combinations[x])
grid_list = sorted(grid_list, key=lambda x: x[0][1])
results_df = {'Input':[], 'Output':[], 'Activation Function':[], 'Hidden Layer':[], 'Epochs':[], 'MAE':[], 'Loss':[]}
for x in range(0,len(grid_list)):
              results_df['Input'].append( grid_list[x][-1][0])
              results_df['Output'].append( grid_list[x][-1][1])
              results_df['Activation Function'].append( grid_list[x][-1][2])
              results_df['Hidden Layer'].append(grid_list[x][-1][3])
              results_df['Epochs'].append(len(grid_search[x][3].epochs))
              results_df['MAE'].append((1-grid_list[x][0][1])*100.0)
              results_df['Loss'].append(grid_list[x][0][0])
results_df = pd.DataFrame(results_df,columns=['Input','Hidden Layer', 'Output', 'Activation Function', 'Epochs','MAE','Loss'])
results_df[:10]
```
